## Immutability

Strings in Python are **immutable** — once created, they cannot be changed. Any operation that appears to modify a string actually creates a **new** string object.

```python
word = 'hello'

# This will raise a TypeError
try:
    word[0] = 'H'
except TypeError as e:
    print(f'Error: {e}')
```

To "change" a string, you create a new one. The original remains unchanged:

```python
word = 'hello'
new_word = 'H' + word[1:]

print(f'original: {word}')
print(f'new:      {new_word}')
```

This is why string methods like `.replace()`, `.upper()`, and `.strip()` always **return a new string** rather than modifying the original. You'll see this pattern throughout the rest of this notebook.

## String Formatting

Python provides three main ways to format strings. Each has its use cases:

### 1. f-strings (Python 3.6+) - **Recommended for most cases**

f-strings are the most readable and fastest method. They allow you to embed expressions directly inside string literals.

```python
lang = 'Python'
version = 3.11

# Basic f-string
secret = f'{lang} is cool'
print(secret)

# With expressions
print(f'{lang} {version} is even better!')

# With formatting
price = 19.99
print(f'Price: ${price:.2f}')  # 2 decimal places
```

### 2. `.format()` method - **Good for templates**

The `.format()` method is useful when you have template strings or need to format the same string multiple times with different values.

```python
# Positional arguments
print('My name is {1} {0}, you can call me {1}.'.format('Doe', 'John'))

# Named arguments (more readable)
print('My name is {first} {family}, you can call me {first}.'.format(first='John', family='Doe'))

# Useful for templates
template = 'Hello {name}, your score is {score}'
print(template.format(name='Alice', score=95))
print(template.format(name='Bob', score=87))
```

### 3. % formatting - **Legacy style (avoid in new code)**

This is the old C-style formatting. It's still supported but f-strings or `.format()` are preferred.

```python
name = 'Python'
# Old style - avoid in new code
message = '%s is %d years old' % (name, 33)
print(message)
```

**When to use which:**
- **f-strings**: Default choice for simple, readable formatting
- **`.format()`**: When using templates or dynamically created format strings
- **`%` formatting**: Only when maintaining legacy code

## Calling multiple methods in a row

```python
ugly_mixed_case = '   ThIS LooKs BAd '
pretty = ugly_mixed_case.strip().lower().replace('bad', 'good')
print(pretty)
```

Note that execution order is from left to right. Thus, this won't work:

```python
pretty = ugly_mixed_case.replace('bad', 'good').strip().lower()
print(pretty)
```

## More Useful String Methods

### `str.startswith()` and `str.endswith()`

```python
filename = 'document.pdf'
print(filename.endswith('.pdf'))  # True
print(filename.startswith('doc'))  # True

url = 'https://www.python.org'
print(url.startswith('https://'))  # True
```

### `str.find()` and `str.index()`

Both find substrings, but `find()` returns -1 if not found, while `index()` raises an exception.

```python
text = 'Python is awesome'
print(text.find('is'))  # Returns position: 7
print(text.find('Java'))  # Returns -1 (not found)

# index() raises ValueError if not found
print(text.index('is'))  # Returns position: 7
# text.index('Java')  # This would raise ValueError
```

### `str.count()`

```python
text = 'how much wood would a woodchuck chuck'
print(text.count('wood'))  # 2
print(text.count('chuck'))  # 2
```

### `str.isdigit()`, `str.isalpha()`, `str.isalnum()`

```python
print('12345'.isdigit())  # True
print('abc123'.isdigit())  # False

print('Python'.isalpha())  # True
print('Python3'.isalpha())  # False

print('Python3'.isalnum())  # True (alphanumeric)
print('Python-3'.isalnum())  # False (contains hyphen)
```
